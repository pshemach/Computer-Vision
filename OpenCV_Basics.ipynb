```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('/content/family.jpg')
```

```python
plt.figure(figsize=(10,8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis(False)
plt.show()
```

```python
text = 'Family'
```

```python
cv2.putText(image,text,(2,30), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,0,255), 4)

plt.figure(figsize=(10,8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis(False)
plt.show()
```

```python
cv2.rectangle(image, (50,50),(500,500),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis(False)
plt.show()
```
